```python
#To-Do-Task
#Dataset Understanding
print("Top 5 rows:")
print(student.head())

print("\nBottom 5 rows:") #\n is for new line for neat output
print(student.tail())

print("\nDataset Info:")
print(student.info())#column, row, int

print("\nDescriptive Statistics:") #mean, median
print(student.describe())

#split Features(X) and Label(Y) .values converts them into NumPy array
X = student[['Math', 'Reading']].values #select math and reading column
Y = student['Writing'].values

```

```python
#To-Do-Task-2 arranges the input data into the correct matric form
X = X.T #transpose swap columns and rows
Y = Y.reshape(-1, 1)
#it makes y a colum vector
```

```python
#To-Do-Task-3
#Train Test Split
def train_test_split_scratch(X, Y, test_size=0.2):
    n = X.shape[1]
    indices = np.random.permutation(n)#randomly shuffle data indices
    test_size = int(n * test_size)#calculate how many saples go to test set

    test_idx = indices[:test_size]
    train_idx = indices[test_size:]

    X_train = X[:, train_idx]
    X_test = X[:, test_idx]
    Y_train = Y[train_idx]
    Y_test = Y[test_idx]

    return X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = train_test_split_scratch(X, Y)

```

```python
#To-Do - 4
#Cost Function(MSE)

def cost_function(X, Y, W):
    n = Y.shape[0]
    Y_pred = np.dot(W.T, X)
    cost = (1 / (2 * n)) * np.sum((Y_pred.T - Y) ** 2)
    return cost

```

```python
#To-Do-5
#Testing a cost function
X_test_case = np.array([[1, 2, 3],
                        [3, 4, 5]])
Y_test_case = np.array([[3], [7], [11]])
W_test_case = np.array([[1], [1]])

cost = cost_function(X_test_case, Y_test_case, W_test_case)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")

print("Cost:", cost)
```

```python
#To-Do-6
#Gradient Descent

def gradient_descent(X, Y, W, alpha, iterations):
    cost_history = []
    m = Y.shape[0]

    for _ in range(iterations):
        Y_pred = np.dot(W.T, X)
        loss = Y_pred.T - Y
        dw = (1 / m) * np.dot(X, loss)
        W = W - alpha * dw
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python
#To-Do-7
#Test code for Gradient Descent function:

np.random.seed(0)
X_test = np.random.rand(3, 100)
Y_test = np.random.rand(100, 1)
W_test = np.random.rand(3, 1)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(X_test, Y_test, W_test, alpha, iterations)

print("Final Parameters:")
print(final_params)

print("Cost History (First 10):")
print(cost_history[:10])

```

```python
#To-Do-8
#RMSE
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))

```

```python
#To-Do-9
#R² Score
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
#To-Do-10
#Main function to Integrate All steps:
#Student Databese

def main():
    student = pd.read_csv("student.csv")

    X = student[['Math', 'Reading']].values.T
    Y = student['Writing'].values.reshape(-1, 1)

    X_train, X_test, Y_train, Y_test = train_test_split_scratch(X, Y)

    W = np.zeros((X_train.shape[0], 1))
    alpha = 0.00001
    iterations = 1000

    W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    Y_pred = np.dot(W_optimal.T, X_test).T

    print("Final Weights:")
    print(W_optimal)

    print("\nCost History (First 10):")
    print(cost_history[:10])

    print("\nRMSE:", rmse(Y_test, Y_pred))
    print("R2 Score:", r2(Y_test, Y_pred))

main()
```

```python
#Problem 1 - Data Read, Write and Inspect:


from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv("medical_students_dataset.csv")
df.head()
```

```python
print("=== DATAFRAME INFO ===")
print(df.info())

print("\n=== MISSING VALUES PER COLUMN ===")
print(df.isnull().sum())

numeric_column = df.select_dtypes(include='number').columns
categorical_col = df.select_dtypes(include='object').columns


df[numeric_column] = df[numeric_column].fillna(df[numeric_column].mean())
print("Filled missing numeric values with MEAN")


df[categorical_col] = df[categorical_col].fillna(df[categorical_col].mode())
print("Filled missing categorical values with MODE")

df.head()
```

```python
import matplotlib.pyplot as plt

# ------------- PROBLEM 1 -------------------
subset = titanic[['Name','Pclass','Sex','Age','Fare','Survived']]
first_class = subset[subset['Pclass'] == 1]

print("\nFare Stats for First Class:")
print("Mean:", first_class['Fare'].mean())
print("Median:", first_class['Fare'].median())
print("Max:", first_class['Fare'].max())
print("Min:", first_class['Fare'].min())

# ------------- PROBLEM 2 -------------------
print("\nMissing Age Values:", subset['Age'].isnull().sum())
subset_no_null_age = subset.dropna(subset=['Age'])

# ------------- PROBLEM 3 -------------------
titanic_encoded = titanic.copy()

one_hot = pd.get_dummies(titanic_encoded['Embarked'], prefix='Embarked')
titanic_encoded = pd.concat([titanic_encoded, one_hot], axis=1)
titanic_encoded.drop(columns=['Embarked'], inplace=True)

print("\nTitanic After One-Hot Encoding:")
print(titanic_encoded.head())

# ------------- PROBLEM 4 -------------------
titanic.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate By Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

# ------------- PROBLEM 5 -------------------
titanic.groupby('Embarked')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate By Port of Embarkation")
plt.xlabel("Embarked")
plt.ylabel("Survival Rate")
plt.show()
```

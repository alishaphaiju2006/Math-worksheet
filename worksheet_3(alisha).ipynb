```python
#Exploratory Data Analysis Practice Exercise - 1.
#Which passenger had the highest fare paid relative to their age?

#Fill missing age + Create fare_per_year
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']

#Subset rows
high_fare_age = titanic[titanic['fare_per_year'] > 5]

#Sort descending
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

#View Name + fare_per_year
result = high_fare_age_srt[['Name', 'fare_per_year']]
result.head()
```

```python
#Adult male with highest fare relative to class

#Create fare_per_class
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']

#Filter adult males
adult_males = titanic[(titanic['Sex'] == 'male') & (titanic['Age'] >= 18)]

#Sort descending
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

#Show Name, Age, fare_per_class
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
result.head()
```

```python
#GROUP-BY PRACTICE
#Percent of total fare revenue per passenger class

#Total fare
total_fare = titanic['Fare'].sum()
total_fare

#Fare per class
fare_class1 = titanic[titanic['Pclass'] == 1]['Fare'].sum()
fare_class2 = titanic[titanic['Pclass'] == 2]['Fare'].sum()
fare_class3 = titanic[titanic['Pclass'] == 3]['Fare'].sum()

#Percentages
revenue_percent = {
    'Class 1': fare_class1 / total_fare * 100,
    'Class 2': fare_class2 / total_fare * 100,
    'Class 3': fare_class3 / total_fare * 100
}

revenue_percent

```

```python
#Q2: Percentage of passengers by age group

#Create age_group column
def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

titanic['age_group'] = titanic['Age'].apply(age_group)

#Total passengers
total_passengers = len(titanic)
total_passengers

#Count each group
group_counts = titanic['age_group'].value_counts()
group_counts

#Convert to percentages
group_percent = (group_counts / total_passengers) * 100
group_percent
```
